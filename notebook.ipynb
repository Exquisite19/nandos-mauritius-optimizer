```python
import sys
from PyQt5.QtWidgets import QDialog, QApplication
from icecreamshopwindow import *  # Assuming this contains the UI class

class MyForm(QDialog):
    def _init_(self):
        super()._init_()
        self.ui = Ui_DialogIceCreamShop()
        self.ui.setupUi(self)

        # Hide the description widgets initially
        self.hideFlavourWidgets()

        # === Connect radio buttons or checkboxes to slots ===
        self.ui.radioButtonFlavours.toggled.connect(self.onRadioButtonToggled)
        self.ui.checkBoxShowFlavours.stateChanged.connect(self.onCheckBoxChanged)

    def hideFlavourWidgets(self):
        self.ui.groupBoxFlavours.setVisible(False)
        self.ui.textEditFlavourDescription.setVisible(False)
        self.ui.labelFlavourDescription.setVisible(False)

    def showFlavourWidgets(self):
        self.ui.groupBoxFlavours.setVisible(True)
        self.ui.textEditFlavourDescription.setVisible(True)
        self.ui.labelFlavourDescription.setVisible(True)

    # === Slots to handle user interaction ===
    def onRadioButtonToggled(self, checked):
        if checked:
            self.showFlavourWidgets()
        else:
            self.hideFlavourWidgets()

    def onCheckBoxChanged(self, state):
        if state == 2:  # Qt.Checked == 2
            self.showFlavourWidgets()
        else:
            self.hideFlavourWidgets()


if __name__ == "_main_":
    app = QApplication(sys.argv)
    w = MyForm()
    w.show()
    sys.exit(app.exec_())
```
